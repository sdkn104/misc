```python
# 出力ファイルを読み込み

import glob
import json
all_results = []
for path in glob.glob("out/*_result.json"):
    with open(path, encoding="utf-8") as f:
        data = json.load(f)
        if isinstance(data, list):
            all_results.extend(data)
        else:
            all_results.append(data)
```
