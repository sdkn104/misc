```python
import seaborn as sns
import matplotlib.pyplot as plt

# fmriデータセットの読み込み
fmri = sns.load_dataset("fmri")

# イベントと部位ごとに時系列のsignal平均を線グラフでプロット
#plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region", ci=None)

plt.title('fMRI信号の時系列プロット')
plt.xlabel('Timepoint (秒)')
plt.ylabel('Signal（ベースラインからの変化量）')
plt.legend(title="Event / Region")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# データのロード
fmri = sns.load_dataset("fmri")

# プロット
g = sns.FacetGrid(fmri, col="subject", col_wrap=5, height=2.5, aspect=1.5)
g.map_dataframe(sns.lineplot, x='timepoint', y='signal', hue='event', style='region', errorbar=None)
g.add_legend()
g.set_axis_labels('timepoint', 'signal')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# データのロード
fmri = sns.load_dataset("fmri")

# subjectごとに平均と標準偏差を計算
stats = fmri.groupby('subject')['signal'].agg(['mean', 'std']).reset_index()

# 棒グラフ＋エラーバーでプロット
plt.figure(figsize=(10, 6))
plt.bar(stats['subject'], stats['mean'], yerr=stats['std'], capsize=5, color='skyblue')
plt.xlabel('Subject')
plt.ylabel('Signal (mean ± std)')
plt.title('Mean and Standard Deviation of Signal per Subject')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
